```python
from threading import Thread, Lock

def square_numbers():
  with lock:
    for i in range(100):
      print(i)

if __name__ == "__main__":
  threads = []
  num_threads = 2
  lock = Lock()
  for i in range(num_threads):
    thread = Thread(target=square_numbers)
    threads.append(thread)
  
  for thread in threads:
    thread.start()

  for thread in threads:
    thread.join()
  
  print('end main')
```
